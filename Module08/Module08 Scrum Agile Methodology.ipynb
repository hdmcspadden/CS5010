### User Stories
a simple form of stating requirements.

As a [user type], I want to [user action] [for a reason]. *for a reason is optional

**Example**: As a casual user, I want to be able to ignore enemies because I want ot be able to put the game down without having to pause or close our of the game.

**Do NOT** say how it is done. Create at high level.

**Usually have acceptance criteria** - how to measure/evaluate the story has been implemented correctly.

User Stories live in Project Backlog.

Order user stories by value.

User Story: who, what, why.

Put self in mind of user.

Clarity around the "who" the specific user is.

### How to follow SCRUM LIGHT

1. Create a product backlog of user stories using as many user types as possible.
2. Select user stories for the first sprint, == first spring backlog, by prioritizing the items in the product backlog. Prrioritization is important because often features and functionality build upon other features.
3. Prioritize selected user stories within sprint backlog.
4. Assign tasks to each of the selected user stories,
5. Team members select tasks they would like to work on.
6. At end of sprint, prototype is ready to show the client and the team receives feedback and thoughts.
7. STeos 2 - 6 are repeated for every subsequent sprint until the product is fully developed and read for delivery.

### Manage Sprint Backlog
1. Any member can add,delete, change sprint backlog
2. Work for the sprint emerges
3. If work is unclear, define a sprint backlog item with larger about of time and break it down later
4. Update work remaining as more becomes know

### A Spring Backlog Is
* A subset of the requirements on the product backlog
* Also, acceptance criteria, task breakdown, and effort estimates

### TEAM COMPOSITION
Everyone in the team is assigned a specific role. Here are some example roles:
 
* Project Manager should ask, “How does this fit into our product?”
* Software Architect should ask, “How will this fit into our code design?”
* User Interface Designer should ask, “How will the user see this?”
* Quality Assurance Manager should ask, “How do we know it doesn’t break something?”

## 8.5 Scrum Process

**What are the benefits of having daily or weekly standup meetings?**
First benefit is the quick airing of impediments to progress, and these impediments are aired to the entire team so that the Scrum Master, and other team members are aware of them, and aware that the team is solving the issue.

A second benefit is that everyone is accountable for making daily progress (or discusing why progress has not been made).

A third benefit is that the meeting is short and tightly managed. It is designed so that people who like to tell long stories (umm, this may be me in many scenarios) can't take over the meeting.

**Why is it important that each team member be accountable to the rest of the team? What happens when they’re not?**
I really enjoyed reading this page on scrum.org about how "Accountability is a Quality of Agile": https://www.scrum.org/resources/blog/accountability-quality-agile. I liked the broader idea of accountability as not only being responsible at the end of a process, but being responsible during a process. As Mr. Verheyen put it "... accountability has less to do with the actual result than it has to do with sharing, explaining and justifying the actions taken or not taken, the decisions made or not made and all associated considerations that have led to the actual result."

Teams where members are not accountable to each other can have poor communication, lack of transparency, which counterintuituvely can lead to a go-it-alone attitude, unhappiness, and probably silo'ed code that won't integrate/work within a larger project.

**What are the benefits of conducting a sprint review or prototype with the client?**
Timely client feedback is priceless. It confirms (or not) that the team is on track, and that the client's goals are still agreed to by the team and the client. The sprint review keeps the client and scrum team accountable to each other to provide value and define that value. The sprint review should be a motivating experience for the client AND the scrum team. The conversation at the sprint review should also help the selection of user stories for the next sprint, so this meeting provides immediate direction.
